### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    'first_name': ['Jason', 'Molly ', 'Tina', 'Jake', 'Amy'],
    'last_name' : ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age' :       [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70],
}

df = pd.DataFrame(data)
print(df)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure(figsize=(8,6))
plt.scatter(
    df['preTestScore'],
    df['postTestScore'],
    s=df['age']*5,
    alpha=0.6,
    c='blue'
)

plt.xlabel('preTestScore', fontsize=12)
plt.ylabel('postTestScore', fontsize=12)
plt.title('preTestScore vs postTestScore (tamanho proporcional Ã  idade)', fontsize=14)


for i, row in df.iterrows():
    plt.text(
        row['preTestScore']+0.5,
        row['postTestScore']+0.5,
        f"{row['first_name']}",
        fontsize=9
    )

plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
df['female'] = [0, 0, 1, 0, 1]


colors = df['female'].map({0: 'blue', 1: 'red'})

plt.figure(figsize=(8,6))
plt.scatter(
    df['preTestScore'],
    df['postTestScore'],
    s=df['postTestScore'] * 4.5,
    c=colors,
    alpha=0.6
)

plt.xlabel('preTestScore', fontsize=12)
plt.ylabel('postTestScore', fontsize=12)
plt.title('preTestScore vs postTestScore\n(Tamanho proporcional Ã  pontuaÃ§Ã£o final e cor pelo sexo)', fontsize=14)


from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Masculino', markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Feminino', markerfacecolor='red', markersize=10)
]
plt.legend(handles=legend_elements, loc='best')

plt.show()
```
